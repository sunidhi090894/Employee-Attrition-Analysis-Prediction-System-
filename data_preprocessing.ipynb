```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder, StandardScaler
```

```python
data = pd.read_excel('employee_data_final.xlsx', sheet_name='employee_data')
data.head()
# print("pretty little baby")
```

```python
data.drop(columns=['HourlyRate', 'MonthlyRate', 'StandardHours', 'DailyRate'], inplace=True, errors='ignore')
data.head()
```

<i>VISUALIZATION</i>

```python
sns.countplot(data=data, x='Attrition')
plt.title("Overall Employee Attrition Count")
plt.xlabel("Attrition Status")
plt.ylabel("Number of Employees")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(data=data, x='Department', hue='Attrition')
plt.title("Attrition by Department")
plt.xlabel("Department")
plt.ylabel("Number of Employees")
plt.xticks(rotation=45)
plt.legend(title="Attrition")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(data=data, y='JobRole', hue='Attrition')
plt.title("Attrition by Job Role")
plt.xlabel("Number of Employees")
plt.ylabel("Job Role")
plt.legend(title="Attrition")
plt.tight_layout()
plt.show()
```
